# 01 - Explore Data
Preview extracted frames and verify labels before training.

```python
from fastai.vision.all import *
import matplotlib.pyplot as plt
```

```python
# Path to extracted dataset
path = Path('../data/dataset')
print('Dataset contents:')
for d in sorted(path.rglob('*')):
    if d.is_dir():
        count = len(list(d.glob('*.png')))
        if count > 0:
            print(f'  {d.relative_to(path)}: {count} images')
```

```python
# Preview sample images from each category
categories = [d.name for d in (path/'train').iterdir() if d.is_dir()]
fig, axes = plt.subplots(len(categories), 5, figsize=(15, 3*len(categories)))

for row, cat in enumerate(sorted(categories)):
    images = list((path/'train'/cat).glob('*.png'))[:5]
    for col, img_path in enumerate(images):
        img = PILImage.create(img_path)
        axes[row, col].imshow(img)
        axes[row, col].set_title(cat if col == 0 else '')
        axes[row, col].axis('off')

plt.tight_layout()
plt.show()
```

```python
# Quick check: load data with fastai and show a batch
dls = ImageDataLoaders.from_folder(
    path, train='train', valid='val',
    item_tfms=Resize(224), bs=16
)
dls.show_batch(max_n=12)
```
