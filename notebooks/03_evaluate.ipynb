# 03 - Evaluate Model
Analyze model performance: confusion matrix, top losses, and test predictions.

```python
from fastai.vision.all import *
```

```python
# Load the trained model
learn = load_learner('../models/action_model.pkl')
```

```python
# Show predictions on validation set
learn.show_results(max_n=12)
```

```python
# Confusion matrix
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix(figsize=(8, 8))
```

```python
# Top losses — images the model got most wrong
interp.plot_top_losses(9, figsize=(15, 10))
```
