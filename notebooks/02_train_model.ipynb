# 02 - Train Model
Train an action classifier using fastai and transfer learning.

**Prerequisites:** Run `python scripts/extract_data.py` first to generate the dataset.

```python
from fastai.vision.all import *
```

```python
# Verify GPU is available
import torch
print(f'PyTorch: {torch.__version__}')
print(f'CUDA available: {torch.cuda.is_available()}')
if torch.cuda.is_available():
    print(f'GPU: {torch.cuda.get_device_name(0)}')
```

```python
# Load dataset
path = Path('../data/dataset')

dls = ImageDataLoaders.from_folder(
    path, train='train', valid='val',
    item_tfms=Resize(224),
    batch_tfms=aug_transforms(),
    bs=32
)

print(f'Classes: {dls.vocab}')
print(f'Training samples: {len(dls.train_ds)}')
print(f'Validation samples: {len(dls.valid_ds)}')
dls.show_batch(max_n=9)
```

```python
# Create learner with pretrained ResNet18
learn = vision_learner(dls, resnet18, metrics=accuracy)
```

```python
# Find optimal learning rate
learn.lr_find()
```

```python
# Train the model
learn.fine_tune(5)
```

```python
# Save the trained model
model_path = Path('../models/action_model.pkl')
learn.export(model_path)
print(f'Model saved to: {model_path}')
```
